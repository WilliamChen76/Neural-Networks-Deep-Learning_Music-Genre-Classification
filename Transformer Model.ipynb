```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from collections import Counter
import torch.nn.functional as F
```

```python
numerical_cols = ['danceability', 'energy', 'loudness', 'speechiness', 'instrumentalness', 'duration_ms']
dummy_cols = ['mode_1', 'language_de', 'language_en', 'language_es', 'language_nl']

# Encoding playlist_genre from 0 to 5
label_encoder = LabelEncoder()
train_df['playlist_genre'] = label_encoder.fit_transform(train_df['playlist_genre'])
test_df['playlist_genre'] = label_encoder.transform(test_df['playlist_genre'])
print("encoded playlist_genre:", train_df['playlist_genre'].values)  # Check
label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print(label_mapping)

# Words count and build words bank
def build_vocab(lyrics):
    all_words = []
    for lyric in lyrics:
        words = lyric.lower().split() 
        all_words.extend(words)
    
    word_counts = Counter(all_words)
    vocab = {word: idx + 1 for idx, (word, _) in enumerate(word_counts.most_common())}  # Starting from 1, 0 for padding
    vocab['<PAD>'] = 0 
    return vocab

# Transform lyrics to token series
def lyrics_to_tokens(lyrics, vocab, max_seq_len):
    tokens = []
    for lyric in lyrics:
        words = lyric.lower().split()
        token_seq = [vocab.get(word, 0) for word in words]  # replacing unknown words with <PAD> 
        if len(token_seq) > max_seq_len:
            token_seq = token_seq[:max_seq_len]  # interrupt
        else:
            token_seq += [0] * (max_seq_len - len(token_seq))  # padding
        tokens.append(token_seq)
    return tokens


MAX_SEQ_LEN = 200
vocab = build_vocab(train_df['lyrics'])
VOCAB_SIZE = len(vocab)
train_df['lyrics'] = lyrics_to_tokens(train_df['lyrics'], vocab, MAX_SEQ_LEN)
test_df['lyrics'] = lyrics_to_tokens(test_df['lyrics'], vocab, MAX_SEQ_LEN)

# Standarsize numerical columns
scaler = StandardScaler()
scaler.fit(train_df[numerical_cols])
train_df[numerical_cols] = scaler.transform(train_df[numerical_cols])
test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])

class MusicDataset(Dataset):
    def __init__(self, df):
        self.lyrics = torch.tensor(np.array(df['lyrics'].tolist()), dtype=torch.long)
        self.numerical = torch.tensor(df[numerical_cols].values, dtype=torch.float32)
        self.dummy = torch.tensor(df[dummy_cols].values, dtype=torch.float32)
        self.labels = torch.tensor(df['playlist_genre'].values, dtype=torch.long)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            'lyrics': self.lyrics[idx],
            'numerical': self.numerical[idx],
            'dummy': self.dummy[idx],
            'label': self.labels[idx]
        }

# # split as train/test
# train_df, val_df = train_test_split(df_final, test_size=0.2, random_state=42)
train_dataset = MusicDataset(train_df)
val_dataset = MusicDataset(test_df)

# DataLoader
train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=2, shuffle=False)
```

```python
class MusicGenreClassifier(nn.Module):
    def __init__(self, vocab_size, max_seq_len, d_model=64, nhead=4, num_layers=2, num_classes=6):
        super(MusicGenreClassifier, self).__init__()

        # Transformer dealing with lyrics
        self.embedding = nn.Embedding(vocab_size, d_model)
        self.pos_encoder = nn.Parameter(torch.zeros(1, max_seq_len, d_model))  # positional encoding
        transformer_layer = nn.TransformerEncoderLayer(d_model=d_model, nhead=nhead) # multi-head self-attention
        self.transformer_encoder = nn.TransformerEncoder(transformer_layer, num_layers=num_layers) # consists of multiple TransformerEncoderLayers
        self.transformer_output_dim = d_model

        # MLP dealing with numerical and dummy columns
        self.numerical_input_dim = len(numerical_cols)  # num
        self.dummy_input_dim = len(dummy_cols)  # dummy
        self.mlp_input_dim = self.numerical_input_dim + self.dummy_input_dim
        self.mlp = nn.Sequential(
            nn.Linear(self.mlp_input_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Dropout(0.1)
        )
        self.mlp_output_dim = 32

        self.final_layer = nn.Linear(self.transformer_output_dim + self.mlp_output_dim, num_classes)

    def forward(self, lyrics, numerical, dummy):
        # Transformer dealing with lyrics
        lyrics_embed = self.embedding(lyrics)  # (batch_size, seq_len, d_model)
        lyrics_embed = lyrics_embed + self.pos_encoder[:, :lyrics.size(1), :]  # positional encoding
        transformer_out = self.transformer_encoder(lyrics_embed.transpose(0, 1))  # (seq_len, batch_size, d_model)
        transformer_out = transformer_out.mean(dim=0)  # avg pooling：(batch_size, d_model)

        # MLP dealing with numerical and dummy columns
        mlp_input = torch.cat((numerical, dummy), dim=1)  # (batch_size, numerical + dummy)
        mlp_out = self.mlp(mlp_input)  # (batch_size, mlp_output_dim)

        # combining outputs of Transformer and MLP 
        combined = torch.cat((transformer_out, mlp_out), dim=1)  # (batch_size, transformer_output_dim + mlp_output_dim)

        # Final classification
        logits = self.final_layer(combined)  # (batch_size, num_classes)
        return logits

# Initialize
model = MusicGenreClassifier(vocab_size=VOCAB_SIZE, max_seq_len=MAX_SEQ_LEN, num_classes=len(label_encoder.classes_))
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
# loss function and optimizer
from sklearn.metrics import confusion_matrix, recall_score, f1_score, accuracy_score
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# train
def train(model, train_loader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for batch in train_loader:
        lyrics = batch['lyrics'].to(device)
        numerical = batch['numerical'].to(device)
        dummy = batch['dummy'].to(device)
        labels = batch['label'].to(device)

        optimizer.zero_grad()
        outputs = model(lyrics, numerical, dummy)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    return total_loss / len(train_loader), correct / total

# validate
def validate(model, val_loader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in val_loader:
            lyrics = batch['lyrics'].to(device)
            numerical = batch['numerical'].to(device)
            dummy = batch['dummy'].to(device)
            labels = batch['label'].to(device)

            outputs = model(lyrics, numerical, dummy)
            loss = criterion(outputs, labels)

            total_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(val_loader)
    accuracy = correct / total
    recall = recall_score(all_labels, all_preds, average='macro')
    f1 = f1_score(all_labels, all_preds, average='macro')
    cm = confusion_matrix(all_labels, all_preds)
    return avg_loss, accuracy, recall, f1, cm

# epochs
num_epochs = 10
best_val_acc = 0.0
best_cm = None
best_model_path = 'best_model.pt'
train_losses = []
train_accuracies = []
val_losses = []
val_accuracies = []
val_recalls = []
val_f1s = []

for epoch in range(num_epochs):
    train_loss, train_acc = train(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc, val_recall, val_f1, cm = validate(model, val_loader, criterion, device)

    train_losses.append(train_loss)
    train_accuracies.append(train_acc)
    val_losses.append(val_loss)
    val_accuracies.append(val_acc)
    val_recalls.append(val_recall)
    val_f1s.append(val_f1)

    print(f'Epoch {epoch+1}/{num_epochs}:')
    print(f'Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}')
    print(f'Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}, Recall: {val_recall:.4f}, F1: {val_f1:.4f}')
    print("Confusion Matrix:\n", cm)

    # Save the model if it has the best validation accuracy
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_cm = cm
        torch.save(model.state_dict(), 'best_model.pt')
        # print(f"Model saved as 'best_model.pt' with validation accuracy: {best_val_acc:.4f}")
        print(f"Model saved as '{best_model_path}' with validation accuracy: {best_val_acc:.4f}")

print("Training completed.")    
```
